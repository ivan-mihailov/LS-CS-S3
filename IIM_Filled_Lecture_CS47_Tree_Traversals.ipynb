```python
class BinarySearchTree:
  def __init__(self, value):
      # the value at the current node
      self.value = value
      # reference to this node's left child
      self.left = None
      # reference to this node's right child
      self.right = None

  def insert(self, value):
      # check if the new node's value is less than our current node's value
      if value < self.value:
          # if there's no left child here already, place the new node here
          if not self.left:
              self.left = BinarySearchTree(value)
          else:
              # otherwise, repeat the process!
              self.left.insert(value)
      # check if the new node's value is greater than or equal to our 
      # current node's value
      elif value >= self.value:
          # if there's no right child here already, place the new node here
          if not self.right:
              self.right = BinarySearchTree(value)
          else:
              # otherwise, repeat the process!
              self.right.insert(value)

  def pre_order_dft(self):
    if self is None:
      return
    # do your thing
    print(self.value)
    # visit the left
    if self.left:
      self.left.pre_order_dft()
    # visit the right
    if self.right:
      self.right.pre_order_dft()

  def in_order_dft(self):
    if self is None:
      return
    # visit the left
    if self.left:
      self.left.in_order_dft() 
    # do your thing
    print(self.value)
    # visit the right
    if self.right:
      self.right.in_order_dft()

  def post_order_dft(self):
    if self is None:
      return
    # visit the left
    if self.left:
      self.left.post_order_dft()
    # visit the right
    if self.right:
      self.right.post_order_dft()
    # do your thing
    print(self.value)
  
  def bft_print(self):
    data = []
    data.append(self)

    while len(data) > 0:
      current = data.pop(0)
      print(current.value)
      if current.left:
        data.append(current.left)
      if current.right:
        data.append(current.right)

  def dft_print(self):
    data = []
    data.append(self)

    while len(data) > 0:
      current = data.pop()
      print(current.value)
      if current.left:
        data.append(current.left)
      if current.right:
        data.append(current.right)

  def for_each(self, cb): # cb(callback function)--passing a function to run over the data
    cb(self.value)
    if self.left:
      self.left.for_each(cb)
    if self.right:
      self.right.for_each(cb)

t = BinarySearchTree('A')
t.left = BinarySearchTree('B')
t.right = BinarySearchTree('C')
t.left.left = BinarySearchTree('D')
t.right.left = BinarySearchTree('E')
t.right.right = BinarySearchTree('F')

t.pre_order_dft()
print()
t.in_order_dft()
print()
t.post_order_dft()
print()
t.bft_print()
print()
t.dft_print()

print()
n = BinarySearchTree(10)

def my_thing(val):
  print(val * 10)

n.for_each(my_thing)
print()  
t.for_each(my_thing) 
```

```python
"""
You are given a binary tree.
Write a function that can return the inorder traversal of node values.
Example:
Input:
   3
    \
     1
    /
   5
Output: [3,5,1]
"""
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def inorder_traversal(root):
    val_list = []
    if root is None:
      return val_list
    if root.left:
      inorder_traversal(root.left)
    val_list.append(root.val)
    if root.right:
      inorder_traversal(root.right)
      #val_list.append(root.right.val)
    return val_list

n = TreeNode(3)
n.right = TreeNode(1)
n.right.left = TreeNode(5)

res = inorder_traversal(n)
print(res)
```
